```python
import numpy as np

def solve(pA,pE):
    qA=1-pA
    qE=1-pE
    p01=pA;p00=1-p01
    p10=qA*pE; p12=pA*qE;  p11=1-p10-p12
    p21=qA*pE; p23= pA*qE; p22=1-p21-p23
    p32=qA*pE; p33=1-qA*pE

    P = [
    [p00,p01,0,0],
    [p10,p11,p12,0],
    [0,p21,p22,p23],
    [0,0,p32,p33]
    ]
    v=[0,0,0,1]
    init=v
    n=0
    for k in range(1,400):
        v = np.dot(v,P)

    return v

x=[]
y=[]
for pa in range(101):
    x.append(pa/100)
    y.append(1-solve(pa/100,0.8)[3])

plt.title("$p_E$ = 0.8")
plt.xlabel("Prob. arrival")
plt.ylabel("Availability")
plt.plot(x,y)
plt.show()
```
