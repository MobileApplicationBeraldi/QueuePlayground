```python
import random
import matplotlib.pyplot as plt
m=0
a=.00005
X=[]
Y=[]
for _ in range(100000):
  x=random.gauss(9,10)
  m=m+a*(x-m)
  X.append(_)
  Y.append(m)


plt.plot(X, Y, label='win')
plt.show()
```
