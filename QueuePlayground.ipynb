```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x_values = np.linspace(0, 0.999, 100)  # Values close to, but not equal to 1 to avoid division by zero

# Calculate the corresponding y values for each function
y1_values = 1 / (1 - x_values)
y2_values = np.log(1 / (1 - x_values))

# Plot both functions on the same set of axes
plt.figure(figsize=(10, 6))

plt.plot(x_values, y1_values, label=r'$\frac{1}{1-x}$')
plt.plot(x_values, y2_values, label=r'$\log\left(\frac{1}{1-x}\right)$')

plt.title('Graphs of Functions')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

def erlang_b(M, A):
    numerator = (A ** M) / factorial(M)
    denominator = sum([(A ** i) / factorial(i) for i in range(M + 1)])
    return numerator / denominator

# Range of traffic intensities (A values)
A_values = np.linspace(0.01, 10, 100)

# Number of servers (M) values
M_values = [1, 5, 10, 20]

# Plot Erlang-B for different values of M
plt.figure(figsize=(10, 6))

for M in M_values:
    B_values = [erlang_b(M, A) for A in A_values]
    plt.plot(A_values, B_values, label=f'M={M}')

plt.title('Erlang-B Blocking Probability')
plt.xlabel('Traffic Intensity (A)')
plt.ylabel('Blocking Probability (B)')
plt.legend()
plt.grid(True)
plt.show()
```
